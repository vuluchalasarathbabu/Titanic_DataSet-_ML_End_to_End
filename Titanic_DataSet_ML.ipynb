**üîç EDA (Exploratory Data Analysis)**

EDA is the process of exploring data before modeling.
We understand the structure, missing values, outliers, and patterns in the dataset.

**Step 1: Data Understanding and Overview**

      1.check how many numerical and categorical features there are.
        Numerical features: Age, SibSp, Parch, Fare
        Categorical features: Sex, Embarked, Pclass, Cabin, Name, Ticket

      2.Check for Missing Values
        You can visualize missing values using all the Graphs
        Example insights:
        Age, Cabin, and Embarked have missing values.

      3.Detect Outliers:
        Visualize with box plots

      4.Clean Raw Data:
        Remove unnecessary columns like Ticket or Cabin (if too many missing values) and prepare the data for analysis.


**Step 2: Handling Missing Values**

  We can replace missing data using:

    Mean (average value)
    Median (middle value)
    Mode (most frequent value)

    Or group-based imputation (fill Age by class or gender)      

**‚öñÔ∏è Step 3: Handling Imbalanced Data**

  Check if the target variable (Survived) is imbalanced:
  If one class dominates, use techniques like:

    Oversampling (SMOTE)

    Undersampling

    Class weights during model training.

**üö´ Step 4: Treating Outliers**

  Outliers are extreme values that can affect model accuracy.
  Treat them by:

    Capping (limit extreme values)

    Transformation (e.g., log transform)

    Removing extreme cases

**‚öôÔ∏è Step 5: Feature Scaling**

  Scaling brings all numerical features to the same scale, which helps algorithms like SVM or KNN.

    Techniques:

    Standardization (Z-score): (x - mean) / std

    Normalization (Min-Max): (x - min) / (max - min)

**üî° Step 6: Encoding Categorical Variables**

  Converting the Categirical Features into numerical features.
  means Convert non-numeric data into numeric form so models can understand.

    Methods:

    Label Encoding: Convert categories into integers

    One-Hot Encoding: Create dummy columns (0/1)


**üß† Feature Engineering**

Feature Engineering means creating new useful features from existing ones.

    Examples on Titanic dataset:

    1. FamilySize = SibSp + Parch + 1
    2. IsAlone = 1 if FamilySize == 1 else 0
    3. Title extracted from Name (Mr, Mrs, Miss, etc.)
    4. Age Groups: Bin ages into Child, Teen, Adult, Senior
    5. Fare Category: Group fares into low, medium, high

These new features often improve model accuracy.

**üßÆ Feature Selection**

Once you have many features, you need to keep only the most important ones that affect survival.

**‚úÖ Step 7: Feature Selection Techniques**

    1Ô∏è‚É£ Correlation Analysis
    Check how strongly features are related.
    Remove highly correlated (redundant) variables.

    2Ô∏è‚É£ K-Nearest Neighbors (K-Best)

    Select top features that have the highest relationship with the target.   

    3Ô∏è‚É£ Chi-Square Test

    Used for categorical data to check dependency between features and target variable.

    4Ô∏è‚É£ Genetic Algorithms

    An advanced optimization method inspired by evolution ‚Äî selects the best combination of features for maximum performance.

    5Ô∏è‚É£ Feature Importance (Tree-Based Models)

    Use algorithms like RandomForest or ExtraTrees to rank features by importance.

**üß© Step 6: Correlation Matrix (Numeric Features)**

    What: Compute pairwise correlations for numeric columns.
    Why: Detect collinearity and features strongly related to the target.

    Correlation measures how two numeric variables move together.

    It‚Äôs a value between ‚Äì1 and +1:

      +1 ‚Üí perfect positive correlation (as one increases, the other increases)

      ‚Äì1 ‚Üí perfect negative correlation (as one increases, the other decreases)

      0 ‚Üí no linear relationship

    In the Titanic dataset, this helps identify which numeric features (like Age, Fare, Pclass) relate to survival.

```
# ---------------------------------------------------------------
# Step 19 ‚Äì Correlation Analysis: Identify Relationships Between Numeric Features
# ---------------------------------------------------------------
# üìò Purpose:
# To examine how numeric variables in the dataset relate to each other.
# Correlation helps identify which features have strong positive or negative associations.
# This is crucial before modeling, as highly correlated variables can cause multicollinearity.
# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Detects strong correlations with the target variable (e.g., Survived).
# ‚Ä¢ Reveals redundant features that might be dropped or combined.
# ‚Ä¢ Provides insight into which numeric features may carry predictive power.
# ---------------------------------------------------------------
# Import necessary visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for correlation computation
# df.select_dtypes(include=['int64', 'float64']) filters out non-numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute and visualize the correlation matrix for numeric features only
plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),     # correlation matrix of numeric columns
    annot=True,            # display correlation coefficients inside cells
    fmt='.2f',             # format with two decimal places
    cmap='coolwarm',       # color gradient: blue = negative, red = positive
    center=0               # center the color scale at 0
)

# Add a title to make the plot clear
plt.title('Numeric Correlation Matrix (Cleaned)')

# Display the heatmap
plt.show()
```

**üß© Step 8 ‚Äî Feature Extraction: Title from Name and Deck from Cabin**

    What: Derive Title and Deck from Name and Cabin.
    Why: Create informative categorical features that capture social/locational signals.

Explaination:

| Column        | What We Do                                                      | Why It‚Äôs Useful                                                                  |
| ------------- | --------------------------------------------------------------- | -------------------------------------------------------------------------------- |
| `Title`       | Extract the word like **Mr, Mrs, Miss, Master, Dr** from `Name` | Adds social context ‚Äî titles often correlate with **age, gender, and survival**. |
| `rare_titles` | Group rare ones under ‚ÄúRare‚Äù                                    | Prevents the model from learning noise from very few samples.                    |
| `Deck`        | Take the **first letter** of `Cabin` (like A‚ÄìG)                 | Indicates **deck location** ‚Äî may relate to survival probability.                |


```

# ---------------------------------------------------------------
# Step 21 ‚Äì Feature Extraction & Data Enrichment
# ---------------------------------------------------------------
# üìò Purpose:
# Create new, meaningful features from existing raw columns to improve model interpretability.
# Extracting titles and deck letters provides valuable insights about social status and cabin location,
# which can correlate with survival probability.
# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Adds socio-economic context via 'Title' (e.g., Mrs, Miss, Mr, Master, Rare)
# ‚Ä¢ Provides cabin/deck spatial information from 'Cabin' data
# ‚Ä¢ Enhances model‚Äôs ability to detect hidden survival patterns
# ---------------------------------------------------------------

# Import NumPy for handling NaN values
import numpy as np

# -------------------------------------------------
# üß† Extract 'Title' from the 'Name' column
# Example: "Braund, Mr. Owen Harris" ‚Üí "Mr"
# -------------------------------------------------
df['Title'] = (
    df['Name']                                     # use the Name column
      .str.extract(r',\s*([^\.]+)\.', expand=False) # regex: find text between ',' and '.'
      .str.strip()                                 # remove extra spaces
)

# -------------------------------------------------
# üßπ Consolidate and clean up similar or rare titles
# -------------------------------------------------
rare_titles = [
    'Lady','Countess','Capt','Col','Don','Dr',
    'Major','Rev','Sir','Jonkheer','Dona'
]

# Replace variants:
# - 'Mlle' and 'Ms' are same as 'Miss'
# - 'Mme' is same as 'Mrs'
df['Title'] = (
    df['Title']
      .replace(['Mlle', 'Ms'], 'Miss')
      .replace(['Mme'], 'Mrs')
)

# Replace rare titles with a single category 'Rare'
df['Title'] = df['Title'].apply(lambda x: 'Rare' if x in rare_titles else x)

# -------------------------------------------------
# üö¢ Extract 'Deck' information from 'Cabin'
# Example: "C85" ‚Üí "C" (the deck letter)
# -------------------------------------------------
# Convert Cabin column to string and take first letter
df['Deck'] = df['Cabin'].astype(str).str[0]

# Replace 'n' (from 'nan' string) with proper NaN
df.loc[df['Deck'] == 'n', 'Deck'] = np.nan

# -------------------------------------------------
# üîé Display sample of extracted columns
# -------------------------------------------------
display(df[['Name', 'Title', 'Cabin', 'Deck']].head())
```

**Step:9 Impute missing values ‚Äî group median for Age (by Title or Sex+Pclass)**

This step fills missing values (NaN) for key columns like Age, Embarked, and Fare using logical strategies based on the dataset.

    What: Fill Age using group medians (e.g., by Title), fill Embarked with mode and Fare with median.
    Why: Replace NaNs with plausible values to enable modeling and reduce bias.

üß† Explanation of Each Step

| Feature      | Strategy                                                                   | Why It‚Äôs Done                                                             |
| ------------ | -------------------------------------------------------------------------- | ------------------------------------------------------------------------- |
| **Age**      | Filled using **median age per Title** (e.g., Miss = 22, Mr = 30, Mrs = 35) | Titles reflect social groups ‚Üí more accurate than using one overall mean. |
| **Embarked** | Filled with **mode** (most common port, usually ‚ÄúS‚Äù)                       | Only 2 missing ‚Üí simple categorical imputation works.                     |
| **Fare**     | Filled with **median fare**                                                | Skewed distribution ‚Üí median is more robust than mean.                    |


```
# ================================================================
# Step 22 ‚Äì Data Cleaning: Impute Missing Values
# ================================================================
# üìò Purpose:
# Handle missing (NaN) values in important columns like Age, Embarked, and Fare.
# Missing data can distort model accuracy, so we impute values based on
# domain knowledge and statistical patterns.
# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Prevents model bias caused by NaN values
# ‚Ä¢ Age imputed intelligently based on social groups (titles)
# ‚Ä¢ Maintains dataset integrity for training machine learning models
# ---------------------------------------------------------------

# Import pandas (if not already imported)
import pandas as pd

# -------------------------------------------------
# üß† AGE: fill missing values intelligently
# -------------------------------------------------
# Create a copy of the 'Age' column to store the imputed version
df['Age_imputed'] = df['Age']

# Compute median age for each Title group (e.g., Mr, Mrs, Miss, Master, etc.)
title_median_age = df.groupby('Title')['Age'].median()

# Compute overall median age (fallback if a title group has no median)
overall_median_age = df['Age'].median()

# Define a function to fill missing ages based on title median
def fill_age(row):
    if pd.isnull(row['Age']):          # If Age is missing
        t = row['Title']               # Get passenger's title
        if pd.notnull(title_median_age.get(t)):   # If median exists for that title
            return title_median_age[t]            # Use title-based median
        else:
            return overall_median_age             # Otherwise use overall median
    else:
        return row['Age']                         # If not missing, keep original value

# Apply the function row by row
df['Age_imputed'] = df.apply(fill_age, axis=1)

# -------------------------------------------------
# ‚öì EMBARKED: fill missing values with mode (most common value)
# -------------------------------------------------
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# -------------------------------------------------
# üí∞ FARE: fill missing values with median
# -------------------------------------------------
df['Fare'] = df['Fare'].fillna(df['Fare'].median())

# -------------------------------------------------
# üìä Report: Check if any missing values remain in key columns
# -------------------------------------------------
display(df[['Age', 'Age_imputed', 'Embarked', 'Fare']].isnull().sum())
```

**Step11: Outliers: visualize & treat (Fare) ‚Äî option to log or cap**

    What: Visualize boxplot and either apply log1p or cap at 99th percentile.
    Why: Reduce the influence of extreme fares so models aren‚Äôt dominated by rare values.
    
    Outliers are extremely high or low values that can distort model performance ‚Äî especially for features like ticket Fare, where a few passengers paid unusually high prices.

```
# ================================================================
# Step 15 ‚Äì Outlier Handling for 'Fare'
# ================================================================
# üéØ Purpose:
# Handle extreme or skewed values in the 'Fare' column, which can
# distort statistical analysis and mislead machine learning models.
# We'll detect, visualize, and fix outliers using two techniques:
# (1) Log Transformation, and (2) Winsorization (Capping).
# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Prevents extreme fares from dominating model training.
# ‚Ä¢ Ensures smoother, more stable model performance.
# ‚Ä¢ Log transformation also helps models sensitive to skewed data (e.g., linear/logistic regression).
# ---------------------------------------------------------------

# Outlier Handling for 'Fare'

# Import required visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1Ô∏è‚É£ Visualize Fare distribution using a boxplot
#    - Boxplot helps to detect outliers (points far above the upper whisker)
plt.figure(figsize=(8,4))
sns.boxplot(df['Fare'].dropna())
plt.title('Fare Boxplot - Detecting Outliers')
plt.show()

# 2Ô∏è‚É£ Option A: Log Transformation (np.log1p handles zeros safely)
#    - Compresses large values and reduces skewness
#    - Makes the data more "normal" (better for some ML algorithms)
df['Fare_log'] = np.log1p(df['Fare'])

# 3Ô∏è‚É£ Option B: Capping Outliers (Winsorization)
#    - We set a limit at the 99th percentile so that
#      extreme high fares are replaced (capped) by this threshold
cap = df['Fare'].quantile(0.99)  # compute 99th percentile value
df['Fare_capped'] = df['Fare'].clip(upper=cap)  # cap all values above 99th percentile

# 4Ô∏è‚É£ Display the capping threshold used
print("Fare 99th percentile cap:", cap)
```

```
# ================================================================
# Step 12 ‚Äì Feature Engineering: Family-related Features
# ================================================================
# üéØ Purpose:
# Enhance the dataset by combining 'SibSp' and 'Parch' to represent
# family relationships. Family size can influence survival chances.
# ---------------------------------------------------------------

# üß† Background:
# ‚Ä¢ 'SibSp' ‚Üí Number of siblings/spouses aboard the Titanic
# ‚Ä¢ 'Parch' ‚Üí Number of parents/children aboard
# Combining them gives insights into family influence on survival.

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Create 'FamilySize'
# ---------------------------------------------------------------
# Formula: FamilySize = SibSp + Parch + 1
# (Adding 1 includes the passenger themselves)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Create 'IsAlone'
# ---------------------------------------------------------------
# Passengers traveling alone (FamilySize == 1) might behave differently
# in emergency situations compared to those with family.
# We convert the Boolean (True/False) to integer (1/0) for ML models.
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

# ---------------------------------------------------------------
# 3Ô∏è‚É£ Preview new engineered features
# ---------------------------------------------------------------
display(df[['SibSp', 'Parch', 'FamilySize', 'IsAlone']].head())

# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Captures family influence on survival probability.
# ‚Ä¢ Adds interpretable, non-linear relationships between features.
# ‚Ä¢ Often improves model accuracy in Titanic-based classification tasks.
# ---------------------------------------------------------------
```

**Step 13: Feature Engineering ‚Äî Binning Continuous Variables (Age & Fare)**

    What: Bin Age into life stages and Fare into quantile-based bins.
    Why: Transform continuous variables into meaningful categories to capture non-linear effects.

```
# ================================================================
# Step 13 ‚Äì Feature Engineering: Age & Fare Binning (Discretization)
# ================================================================
# üéØ Purpose:
# Transform continuous variables ('Age' and 'Fare') into categorical bins.
# This helps capture non-linear relationships and can improve interpretability
# for both EDA and machine learning models.
# ---------------------------------------------------------------

import pandas as pd

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Age Binning
# ---------------------------------------------------------------
# ‚Ä¢ 'Age_imputed' is divided into 5 logical life-stage categories.
# ‚Ä¢ Binning helps reveal patterns such as higher survival among children.
# ‚Ä¢ pd.cut() creates equal-width bins using numeric boundaries.
df['AgeBin'] = pd.cut(
    df['Age_imputed'],
    bins=[0, 12, 18, 35, 60, 120],
    labels=['Child', 'Teen', 'Young', 'Adult', 'Senior']
)

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Fare Binning
# ---------------------------------------------------------------
# ‚Ä¢ 'Fare_capped' is divided into 4 quartiles (equal-frequency bins).
# ‚Ä¢ pd.qcut() ensures each bin has roughly equal number of passengers.
# ‚Ä¢ This transformation handles skewed fare distributions effectively.
df['FareBin'] = pd.qcut(
    df['Fare_capped'],
    q=4,
    labels=['VeryLow', 'Low', 'High', 'VeryHigh']
)

# ---------------------------------------------------------------
# 3Ô∏è‚É£ Verify New Features
# ---------------------------------------------------------------
display(df[['Age_imputed', 'AgeBin', 'Fare_capped', 'FareBin']].head())

# ---------------------------------------------------------------
# ‚úÖ Why this step matters:
# ‚Ä¢ Simplifies complex numeric features into intuitive categories.
# ‚Ä¢ Highlights patterns (e.g., children or low-fare passengers‚Äô survival).
# ‚Ä¢ Enhances performance for tree-based or categorical ML models.
# ---------------------------------------------------------------
```

**üß© Step 14 ‚Äî Prepare Features for Modeling (Encoding & Selection)**

  **Prepare features for modeling: drop columns, encode categorical**

    What: Choose features, drop identifiers, and one-hot encode categoricals.
    Why: Produce a numeric feature matrix X and target y suitable for ML models.

```
# ================================================================
# Step 14 ‚Äì üß© Prepare Dataset for Machine Learning
# ================================================================
# üéØ Purpose:
# Before training models, we must:
#   1Ô∏è‚É£ Select relevant columns (features)
#   2Ô∏è‚É£ Drop unused or redundant columns
#   3Ô∏è‚É£ Convert categorical variables into numeric form
#   4Ô∏è‚É£ Separate input features (X) and target (y)
# ---------------------------------------------------------------
# ‚úÖ Outcome:
# X ‚Üí Clean, numeric feature matrix ready for ML models
# y ‚Üí Target variable (Survived)
# ---------------------------------------------------------------

import pandas as pd

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Drop Unnecessary Columns
# ---------------------------------------------------------------
# These columns are either identifiers or replaced by engineered features.
drop_cols = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'Age', 'Fare']
df = df.drop(columns=drop_cols, errors='ignore')

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Define Relevant Features
# ---------------------------------------------------------------
# These were identified through EDA and feature engineering steps.
features = [
    'Pclass', 'Sex', 'Age_imputed', 'Fare_capped', 'SibSp', 'Parch',
    'FamilySize', 'IsAlone', 'Embarked', 'Title', 'Deck', 'AgeBin', 'FareBin'
]

# ---------------------------------------------------------------
# 3Ô∏è‚É£ Safety Check
# ---------------------------------------------------------------
# Keep only those columns that actually exist in df.
features = [f for f in features if f in df.columns]

# ---------------------------------------------------------------
# 4Ô∏è‚É£ Split into Features (X) and Target (y)
# ---------------------------------------------------------------
X = df[features].copy()
y = df['Survived'].copy()

# ---------------------------------------------------------------
# 5Ô∏è‚É£ Encode Categorical Features
# ---------------------------------------------------------------
# Convert all categorical variables to numeric using one-hot encoding.
# drop_first=True removes redundant columns to avoid multicollinearity.
X = pd.get_dummies(X, drop_first=True)

# ---------------------------------------------------------------
# 6Ô∏è‚É£ Quick Check
# ---------------------------------------------------------------
print("‚úÖ Feature matrix shape:", X.shape)
display(X.head())

# ---------------------------------------------------------------
# üí° Why this step matters:
# ‚Ä¢ Transforms mixed data (text + numbers) into fully numeric form.
# ‚Ä¢ Removes noise and redundancy from the dataset.
# ‚Ä¢ Produces clean, ready-to-train ML inputs for model building.
# ---------------------------------------------------------------
```

**üß© Step 15 ‚Äî Scale Numeric Columns (StandardScaler) ‚Äî keep pipeline-friendly**

    What: Standardize numeric features to zero mean and unit variance.
    Why: Ensure features contribute comparably and help many algorithms converge.

```
# ================================================================
# Step 15 ‚Äì ‚öñÔ∏è Feature Scaling (Standardization)
# ================================================================
# üéØ Purpose:
# Many machine learning models (e.g., Logistic Regression, SVM, KNN, Neural Networks)
# are sensitive to the **scale** of input features.
#
# ‚û§ Example:
#     - 'Fare' might range in hundreds,
#     - while 'SibSp' only goes from 0‚Äì8.
#
# Without scaling, larger values dominate smaller ones,
# skewing the learning process.
#
# ‚úÖ Standardization rescales features so that:
#     mean = 0  and  standard deviation = 1
# ---------------------------------------------------------------

from sklearn.preprocessing import StandardScaler

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Select Numeric Features for Scaling
# ---------------------------------------------------------------
numeric_features = ['Age_imputed', 'Fare_capped', 'SibSp', 'Parch', 'FamilySize']

# Keep only those columns that exist in X (safe practice)
numeric_features = [col for col in numeric_features if col in X.columns]

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Initialize Scaler
# ---------------------------------------------------------------
# StandardScaler subtracts the mean and divides by the standard deviation.
scaler = StandardScaler()

# ---------------------------------------------------------------
# 3Ô∏è‚É£ Fit & Transform
# ---------------------------------------------------------------
# Fit the scaler on numeric columns and replace them with scaled values.
# ‚ö†Ô∏è Note: In real ML pipelines, fit() is done on training data only.
X[numeric_features] = scaler.fit_transform(X[numeric_features])

# ---------------------------------------------------------------
# 4Ô∏è‚É£ Verify Scaling Results
# ---------------------------------------------------------------
# After standardization, each feature should have mean ‚âà 0 and std ‚âà 1.
scaled_summary = X[numeric_features].describe().loc[['mean', 'std']]
display(scaled_summary)

# ---------------------------------------------------------------
# üí° Why this step matters:
# ‚Ä¢ Prevents domination by large-scale features.
# ‚Ä¢ Speeds up gradient-based model convergence.
# ‚Ä¢ Improves distance-based model accuracy (KNN, SVM, etc.).
# ---------------------------------------------------------------
```

**‚öñÔ∏è Step 17 ‚Äî Handle Imbalance (try SMOTE or Random Oversampling)**

    What: Apply SMOTE to synthesize minority samples or fallback to random oversampling.
    Why: Balance classes in training data to prevent biased predictions toward the majority.

    Your training data (X_train_res, y_train_res) is now balanced, ready for model training ‚Äî avoiding bias toward the majority class.

| Method                  | Description                                                 | When to Use                                       |
| ----------------------- | ----------------------------------------------------------- | ------------------------------------------------- |
| **SMOTE**               | Synthesizes new samples between existing minority instances | Preferred when features are continuous            |
| **Random Oversampling** | Simply duplicates minority samples                          | Useful fallback when SMOTE fails or data is small |


```
# ================================================================
# Step 18 ‚Äì ‚öñÔ∏è Handle Imbalance in Target Variable (Survived)
# ================================================================
# üéØ Purpose:
# In many real-world datasets (including Titanic),
# the target classes (Survived = 0 / 1) are *not equally represented*.
#
# Example:
#   Survived = 0 ‚Üí 62% (majority)
#   Survived = 1 ‚Üí 38% (minority)
#
# ‚ûú Problem:
# Machine Learning models may become *biased* toward the majority class,
# predicting ‚ÄúNot Survived‚Äù more often.
#
# ‚ûú Solution:
# We rebalance the dataset using one of two techniques:
#   üîπ Option A: SMOTE (Synthetic Minority Oversampling Technique)
#   üîπ Option B: Random Oversampling (fallback if SMOTE unavailable)
# ---------------------------------------------------------------

try:
    # ---------------------------------------------------------------
    # 1Ô∏è‚É£ SMOTE ‚Äì Generate Synthetic Minority Samples
    # ---------------------------------------------------------------
    from imblearn.over_sampling import SMOTE

    sm = SMOTE(random_state=42)
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

    print("‚úÖ SMOTE applied successfully.")
    print("New class distribution:", y_train_res.value_counts().to_dict())

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Fallback ‚Äì Random Oversampling (if SMOTE fails)
# ---------------------------------------------------------------
except Exception as e:
    print("‚ö†Ô∏è SMOTE unavailable or failed:", e)
    print("‚û°Ô∏è Applying Random Oversampling instead...")

    from sklearn.utils import resample

    # Combine X_train and y_train temporarily
    Xy = pd.concat([X_train, y_train], axis=1)

    # Separate majority and minority classes
    majority = Xy[Xy['Survived'] == 0]
    minority = Xy[Xy['Survived'] == 1]

    # Upsample minority class
    minority_upsampled = resample(
        minority,
        replace=True,
        n_samples=len(majority),  # match majority count
        random_state=42
    )

    # Combine back into balanced dataset
    Xy_upsampled = pd.concat([majority, minority_upsampled])

    # Split back into X and y
    X_train_res = Xy_upsampled.drop('Survived', axis=1)
    y_train_res = Xy_upsampled['Survived']

    print("‚úÖ Random Oversampling applied successfully.")
    print("New class distribution:", y_train_res.value_counts().to_dict())

# ---------------------------------------------------------------
# üí° Why this step matters:
# ‚Ä¢ Prevents model bias toward dominant class.
# ‚Ä¢ Improves recall (sensitivity) for minority class (Survived=1).
# ‚Ä¢ Leads to fairer and more generalizable predictions.
# ---------------------------------------------------------------
```

```
# ================================================================
# Step 18 ‚Äì üéØ Feature Selection using Multiple Methods
# ================================================================
# üéØ Purpose:
# Identify which input features contribute most to predicting survival (target = Survived)
#
# Why this matters:
# - Reduces model complexity (removes noisy or redundant variables)
# - Improves model interpretability
# - Speeds up training and prevents overfitting
# ---------------------------------------------------------------
# Methods used:
#   1Ô∏è‚É£ Correlation Matrix  ‚Üí detects linear relationships
#   2Ô∏è‚É£ SelectKBest (Chi-Square Test) ‚Üí statistical relevance for categorical/encoded data
#   3Ô∏è‚É£ ExtraTreesClassifier ‚Üí model-based importance using ensemble trees
# ================================================================

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Correlation Heatmap ‚Äì Quick Overview
# ---------------------------------------------------------------
plt.figure(figsize=(12, 10))
corr = pd.concat([X, y], axis=1).corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix (Including Target)')
plt.show()

# ---------------------------------------------------------------
# 2Ô∏è‚É£ SelectKBest (Chi-Square Test)
# ---------------------------------------------------------------
# The Chi2 test works only on non-negative values,
# so we scale features to a 0‚Äì1 range first.
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X_mms = pd.DataFrame(mms.fit_transform(X), columns=X.columns)

# Apply Chi2 test to select top K features
k_value = min(10, X_mms.shape[1])   # limit to 10 or total number of features
skb = SelectKBest(score_func=chi2, k=k_value)
skb.fit(X_mms, y)

# Retrieve and display feature scores
chi2_scores = pd.Series(skb.scores_, index=X.columns).sort_values(ascending=False)
print("üîπ SelectKBest (Chi2) ‚Äì Top Ranked Features:")
display(chi2_scores.head(15))

# ---------------------------------------------------------------
# 3Ô∏è‚É£ ExtraTreesClassifier ‚Äì Tree-Based Feature Importance
# ---------------------------------------------------------------
from sklearn.ensemble import ExtraTreesClassifier

# Train an ExtraTrees model to assess feature importance
etc = ExtraTreesClassifier(n_estimators=200, random_state=42)
etc.fit(X, y)

# Extract feature importances
tree_importances = pd.Series(etc.feature_importances_, index=X.columns).sort_values(ascending=False)

print("üå≤ ExtraTrees ‚Äì Top Feature Importances:")
display(tree_importances.head(20))

# ---------------------------------------------------------------
# üí° Insights:
# ‚Ä¢ Features appearing consistently in top positions (across methods)
#   are strong predictors of survival (e.g., Sex, Pclass, Fare, Title).
# ‚Ä¢ Weak features may be dropped to simplify the model later.
# ---------------------------------------------------------------
```

**üß† Step 19 ‚Äî Optional: Recursive Feature Elimination (RFE)**

    What: Iteratively remove least-important features using a chosen estimator.
    Why: Find a compact subset of features that delivers similar or better performance.

    RFE works by recursively fitting a model, ranking features by importance, and removing the weakest features until only the top N remain.
    Here, it uses Logistic Regression to select the 8 most predictive features for the Titanic dataset.

```
# ---------------------------------------------------------------
# Step 19 - Recursive Feature Elimination (RFE)
# ---------------------------------------------------------------
# üéØ Objective:
# To automatically find the most important subset of features for predicting 'Survived'
#
# üß† How it works:
# RFE (Recursive Feature Elimination) repeatedly:
#   1Ô∏è‚É£ Trains a model (here Logistic Regression)
#   2Ô∏è‚É£ Evaluates feature importance (coefficients)
#   3Ô∏è‚É£ Removes the least important feature(s)
#   4Ô∏è‚É£ Repeats until only the desired number of features remains
#
# ‚úÖ Why use RFE?
# It‚Äôs model-based feature selection ‚Üí keeps only features that
# contribute most to prediction accuracy, reducing noise & overfitting.
# ---------------------------------------------------------------

# Import required libraries
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Step 1Ô∏è‚É£: Define base estimator (model to guide feature importance)
# Logistic Regression is simple and interpretable.
lr = LogisticRegression(max_iter=1000, random_state=42)

# Step 2Ô∏è‚É£: Initialize the RFE selector
# n_features_to_select=8 ‚Üí we want the top 8 most predictive features
# step=1 ‚Üí removes one feature per iteration
rfe = RFE(estimator=lr, n_features_to_select=8, step=1)

# Step 3Ô∏è‚É£: Fit the RFE object on the full feature dataset (X, y)
# This will recursively eliminate less important columns.
rfe.fit(X, y)

# Step 4Ô∏è‚É£: Get Boolean mask of selected features (True = selected)
rfe_support = pd.Series(rfe.support_, index=X.columns)

# Step 5Ô∏è‚É£: Print and display final selected features
print("‚úÖ RFE Selected Features:")
selected_features = rfe_support[rfe_support == True].index.tolist()
display(selected_features)

# ---------------------------------------------------------------
# (Optional) Step 6Ô∏è‚É£: View ranking of *all* features
# Lower rank = more important (rank 1 means selected)
# ---------------------------------------------------------------
rfe_ranking = pd.Series(rfe.ranking_, index=X.columns)
print("\nüìä Full RFE Feature Ranking (1 = Selected):")
display(rfe_ranking.sort_values())

# ---------------------------------------------------------------
# (Optional) Step 7Ô∏è‚É£: Evaluate model performance using selected features
# ---------------------------------------------------------------
# Create a reduced feature dataset
X_selected = X[selected_features]

# Retrain the Logistic Regression model with only selected features
lr.fit(X_selected, y)

# Print training accuracy (for reference)
print("\nüéØ Model accuracy using selected features:", lr.score(X_selected, y))
#üßæ Output Interpretation:
#selected_features ‚Üí Top 8 features RFE kept.
#rfe_ranking ‚Üí Shows relative importance (1 = most important).
#Final Accuracy ‚Üí Helps check if model performance improved or remained stable with fewer features
```

**üéØ Step 22 ‚Äî Hyperparameter Tuning with GridSearchCV (Random Forest)**

    What: Search parameter grid with cross-validation to find best model settings.
    Why: Improve model performance and robustness by tuning key hyperparameters.

    üß† Explanation:

    GridSearchCV systematically tests every combination of parameters from param_grid.

    Uses cross-validation (cv=5) to ensure results are not due to random data splits.

    best_estimator_ ‚Üí gives the model with the best accuracy score.

    This step helps prevent overfitting and ensures your model is optimized for unseen data.



```
# ---------------------------------------------------------------
# Step 22 - Hyperparameter Tuning (GridSearchCV) for RandomForest
# ---------------------------------------------------------------
# üéØ Goal:
# Optimize the Random Forest model by finding the best hyperparameters
# using GridSearchCV ‚Äî which systematically tests parameter combinations
# via cross-validation to improve accuracy and generalization.
# ---------------------------------------------------------------

# ‚úÖ Why this step matters:
# - Default Random Forest settings may not be optimal for your dataset.
# - Tuning improves performance by balancing bias vs variance.
# - Cross-validation ensures parameters generalize well to unseen data.
# ---------------------------------------------------------------

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# ---------------------------------------------------------------
# 1Ô∏è‚É£ Define Hyperparameter Search Space
# ---------------------------------------------------------------
# These parameters control the complexity and performance of the forest:
# - n_estimators: Number of decision trees in the ensemble
# - max_depth: Maximum depth of each tree (None = grow fully)
# - min_samples_split: Minimum samples required to split a node
param_grid = {
    'n_estimators': [100, 200],   # try 100 and 200 trees
    'max_depth': [None, 5, 10],   # shallow to deep trees
    'min_samples_split': [2, 5],  # stricter splits
}

# ---------------------------------------------------------------
# 2Ô∏è‚É£ Initialize Base Random Forest
# ---------------------------------------------------------------
# random_state=42 ensures reproducible results
rf = RandomForestClassifier(random_state=42)

# ---------------------------------------------------------------
# 3Ô∏è‚É£ Setup GridSearchCV
# ---------------------------------------------------------------
# - cv=5 ‚Üí 5-fold cross-validation (each model trained 5 times)
# - scoring='accuracy' ‚Üí optimize for accuracy metric
# - n_jobs=-1 ‚Üí parallelize using all available CPU cores
gs = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# ---------------------------------------------------------------
# 4Ô∏è‚É£ Fit GridSearch on Training Data
# ---------------------------------------------------------------
# Use resampled data (balanced) if available, else original training set
gs.fit(X_tr, y_tr)

# ---------------------------------------------------------------
# 5Ô∏è‚É£ Retrieve Best Parameters & Model
# ---------------------------------------------------------------
print("‚úÖ Best Hyperparameters Found:")
print(gs.best_params_)   # Display the best parameter combination

# Extract the best model for evaluation
best_rf = gs.best_estimator_

# ---------------------------------------------------------------
# 6Ô∏è‚É£ Evaluate the Tuned Model on Test Data
# ---------------------------------------------------------------
y_pred_best = best_rf.predict(X_test)

print("\nüìä Tuned Random Forest Model Evaluation:")
print("Accuracy:", round(accuracy_score(y_test, y_pred_best), 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))

# Optional: Show feature importance from tuned model
import pandas as pd
feat_imp = pd.Series(best_rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nüå≤ Top 10 Important Features:")
display(feat_imp.head(10))
```

```
# ---------------------------------------------------------------
# Step 23 ‚Äì Cross-Validation to check model stability
# ---------------------------------------------------------------
# What & Why:
# Even though we already evaluated our tuned Random Forest model on a single
# train-test split, that gives just one accuracy estimate.
#
# ‚û§ Problem: A single split might be "lucky" or "unlucky" ‚Äî it can overestimate
# or underestimate the model‚Äôs true performance.
#
# ‚û§ Solution: Cross-validation (CV) trains and tests the model on multiple
# different splits of the data, giving a *more reliable and stable* measure of performance.
#
# Here, we‚Äôll use 5-Fold CV ‚Üí the dataset is split into 5 parts:
# - 4 folds used for training, 1 for testing, repeated 5 times.
# - The mean accuracy across folds shows how consistent the model is.
# ---------------------------------------------------------------

from sklearn.model_selection import cross_val_score  # used to perform cross-validation

# Perform 5-fold cross-validation on the entire dataset (X, y)
# best_rf = tuned Random Forest model from previous step
# scoring='accuracy' ‚Üí we measure accuracy score on each fold
cv_scores = cross_val_score(best_rf, X, y, cv=5, scoring='accuracy')

# Print the individual accuracy scores for each fold
print("Cross-Validation Scores:", cv_scores)

# Calculate and print the mean accuracy (average performance)
print("Mean CV Accuracy:", cv_scores.mean().round(4))

# Calculate and print the standard deviation (how much scores vary across folds)
# ‚û§ Low std ‚Üí model is stable and generalizes well
# ‚û§ High std ‚Üí model is unstable, performance fluctuates depending on data split
print("Std Deviation:", cv_scores.std().round(4))
#Why this matters:
#‚úÖ Cross-validation ensures that your model performs consistently across different subsets of data.
#üìâ It helps detect overfitting ‚Äî if accuracy drops significantly between folds, your model might be too specific to certain samples.
#üìà The mean score gives an unbiased estimate of generalization, while standard deviation reveals model stability.
```

**‚öôÔ∏è Step 25 ‚Äì Pipeline Integration (Scaling + Model)**


    Why This Matters:

    End-to-End Consistency:
    The same transformations used during training are automatically applied at prediction time ‚Äî no manual scaling errors.

    Cleaner & Reusable Code:
    The pipeline wraps preprocessing and model training in one object ‚Üí simpler to deploy and maintain.

    Cross-Validation Ready:
    Pipelines integrate seamlessly with GridSearchCV or cross_val_score, ensuring scaling happens within each fold (avoiding data leakage).

    Deployment Simplicity:
    You can save this entire pipeline (pipeline.pkl) ‚Äî one file handles everything (scaling + model inference).

```
# ---------------------------------------------------------------
# ‚öôÔ∏è  Step 25 ‚Äì Pipeline Integration (Scaling + Model)
# ---------------------------------------------------------------
# What & Why:
# Combine preprocessing (like scaling) and model training
# into a single unified pipeline. This ensures consistent
# data handling during both training and prediction.
# ---------------------------------------------------------------

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# ---------------- Create a Pipeline ----------------
# The pipeline automatically executes each step in sequence:
#   1Ô∏è‚É£ Apply StandardScaler to normalize numeric features.
#   2Ô∏è‚É£ Train a Random Forest classifier on the scaled data.
#
# Benefit: You never forget to scale during inference ‚Äî the pipeline handles it.
pipeline = Pipeline([
    ('scaler', StandardScaler()),                     # Step 1: Feature Scaling
    ('rf', RandomForestClassifier(random_state=42))    # Step 2: Model Training
])

# ---------------- Train the Pipeline ----------------
# Fit() will:
#   ‚Üí Fit the scaler on X_train
#   ‚Üí Transform X_train
#   ‚Üí Fit the Random Forest on the transformed data
pipeline.fit(X_train, y_train)

# ---------------- Make Predictions ----------------
# predict() will:
#   ‚Üí Apply the same scaling to X_test (using the scaler fitted on X_train)
#   ‚Üí Then predict using the trained Random Forest model
y_pred_pipe = pipeline.predict(X_test)

# ---------------- Evaluate Performance ----------------
acc = accuracy_score(y_test, y_pred_pipe)
print(f"Pipeline RF Accuracy: {acc:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_pipe))
```

```
# ---------------------------------------------------------------
# Step 28 ‚Äì Compare Baseline and Tuned Models
# ---------------------------------------------------------------
# üéØ What & Why:
# After training multiple models (Logistic Regression, Random Forest, etc.),
# it's important to compare their performance side-by-side.
# This helps identify which model performs best overall
# in terms of predictive accuracy and F1-Score.
# ---------------------------------------------------------------

import pandas as pd
from sklearn.metrics import accuracy_score, f1_score

# ‚úÖ 1Ô∏è‚É£ Create a DataFrame summarizing model performance
# Each model‚Äôs predictions (y_pred_lr, y_pred_rf, etc.) are compared against the true labels (y_test)
# We compute both Accuracy and F1-score for fairness:
#  - Accuracy ‚Üí % of correct predictions
#  - F1-score ‚Üí balances precision and recall, useful when classes are imbalanced
results = pd.DataFrame({
    'Model': [
        'Logistic Regression',  # Baseline model (simple linear model)
        'Random Forest',        # Default random forest model
        'Tuned RF',             # Random forest after hyperparameter tuning
        'Pipeline RF'           # Random forest wrapped in a preprocessing pipeline
    ],

    # Accuracy for each model
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr),   # Logistic Regression
        accuracy_score(y_test, y_pred_rf),   # Random Forest (default)
        accuracy_score(y_test, y_pred_best), # Tuned Random Forest
        accuracy_score(y_test, y_pred_pipe)  # Pipeline Random Forest
    ],

    # F1-score for each model
    'F1-Score': [
        f1_score(y_test, y_pred_lr),   # Logistic Regression
        f1_score(y_test, y_pred_rf),   # Random Forest (default)
        f1_score(y_test, y_pred_best), # Tuned Random Forest
        f1_score(y_test, y_pred_pipe)  # Pipeline Random Forest
    ]
})

# ‚úÖ 2Ô∏è‚É£ Sort models by accuracy (descending)
# Makes it easy to visually identify which model performed best
results.sort_values(by='Accuracy', ascending=False, inplace=True)

# ‚úÖ 3Ô∏è‚É£ Display the final comparison table
# Jupyter-friendly view showing all metrics neatly
display(results)

# ‚úÖ 4Ô∏è‚É£ Identify the best model based on Accuracy
best_model = results.iloc[0]
print(f"\nüèÜ Best Performing Model: {best_model['Model']}")
print(f"Accuracy: {best_model['Accuracy']:.4f} | F1-Score: {best_model['F1-Score']:.4f}")

# ---------------------------------------------------------------
# üìò Interpretation:
# - The model with the highest Accuracy & F1-Score is generally your best performer.
# - A tuned Random Forest often outperforms the baseline Logistic Regression,
#   showing how hyperparameter optimization improves predictive power.
# - Pipeline RF ensures end-to-end preprocessing (e.g., scaling, imputation)
#   is integrated, making it robust and production-ready.
# ---------------------------------------------------------------

# üí° Why This Matters:
# Comparing models quantitatively helps you:
#   ‚úî Select the best model for deployment.
#   ‚úî Avoid bias toward complex models without measurable gain.
#   ‚úî Understand trade-offs between simplicity and performance.
# ---------------------------------------------------------------
```
